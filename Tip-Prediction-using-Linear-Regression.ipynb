```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import sklearn
```

```python
import statsmodels
```

```python
from sklearn import linear_model
```

```python
Model = linear_model.LinearRegression()
```

```python
df_tips = sns.load_dataset('tips')
```

```python
df_tips.head()
```

```python
df_tips.corr()
```

```python
X = df_tips[['total_bill']] 
y = df_tips['tip']
```

```python
Model.fit(X, y)
```

```python
Model.predict(X)
```

```python
df_tips['Tip_prediction'] = Model.predict(X)
```

```python
df_tips
```

```python
df_tips['Error'] = abs(df_tips['tip'] - df_tips['Tip_prediction'])
```

```python
df_tips
```

```python
sns.displot(df_tips['Error'], bins=15, kde=True)
plt.title("Error Histogram")
plt.show()
```

```python
sns.boxplot(df_tips['Error'])
plt.title('Total Bill Boxplot')
plt.show()
```
